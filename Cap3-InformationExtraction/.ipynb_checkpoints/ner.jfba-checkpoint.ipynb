```python
import es_core_news_sm  # procesador de lenguaje con segmentación de frases, tokenización, lematización, etiquetas POS y dependencias sintácticas. Carga el modelo en español de Spacy
from spacy import displacy # biblioteca gráfica de Spacy para visualizar 
            # 1) árboles de dependencias ( style ="dep")
            # 2) entidades (style = "ent") -> queremos esto
```

```python
nlp = es_core_news_sm.load()
```

```python
def ExtraerEntidades(texto):
   doc = nlp(texto) 
   entities = [NE for NE in doc.ents] # Cada NE (Named Entity) es un objeto de entidad de spaCy con atributos como: NE.text → el texto de la entidad (“Ministerio de Sanidad”) y NE.label_ → la etiqueta NER (“ORG”, “LOC”, etc.)
   return(entities) #Crea una lista entities con todas esas entidades.
```

```python
def FiltrarEntidades(Entidades, tipo_entidad):
   entidades = list()
   for Ent in Entidades:
        if (Ent.label_ == tipo_entidad):
            entidades.append(Ent.text)
   return(entidades)
```

```python
FILENAME = r'D:\Repos Github\AnaliticaTextual-Ejemplos\CORPUS\deportes\d20.txt'
```

```python
texto = open(FILENAME, 'r',encoding="utf-8").read()
texto.rstrip("\n")
```

```python
entidades= ExtraerEntidades(texto)
print(entidades)
```

```python
entidadesTipo = FiltrarEntidades(entidades,'ORG')
print(entidadesTipo)
```

```python
# La visualización se puede observar en: http://localhost:5000/

doc = nlp(texto)
displacy.serve(doc, style="ent")
```
