```python
#Función para leer un texto o documentos

def LeerTexto(FileName):
    with open(FileName, 'r', encoding='latin-1') as f:
        texto = f.read().split('.') #Parte el texto en una lista, usando el punto como separador.
    return(texto)
```

```python
# Función que realiza el preprocesamiento 

def PreProcesamiento(oraciones):
    texto_limpio = []
    for texto in oraciones:  
       texto = EliminarStopwords(texto)    
       texto = EliminaNumeroYPuntuacion(texto) # elimina caracteres especiales y números      
       if len(texto)!=0:
          texto = regex.sub(' +', ' ', texto) # cambia espacios multiples por espacios simples
          texto_limpio.append(texto)
    return(texto_limpio)
```
